```python
import pandas as pd 
df = pd.read_excel("Dataset.xlsx")
display(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
height_column = df['Height']
df['Height'] = np.random.randint(150,181,size=len(df))
w = df["Salary"].mean()
x = df["Number"].mean()
y = df["Age"].mean()
z = df["Weight"].mean()
df.fillna({"Salary":w}, inplace = True)
df.fillna({"Number":x}, inplace = True)
df.fillna({"Age":y}, inplace = True)
df.fillna({"Weight":z}, inplace = True)
df.fillna({"College":'Community College'}, inplace = True)
df["Salary"] = df["Salary"].astype(int)
display(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
a = df["Team"].value_counts()
b = df["Team"].count()
c = (a/b)*100
t = (df["Team"]).unique()
display(c)
plt.bar(t,c)
plt.title('Percentage Split')
plt.xlabel('Team')
plt.ylabel('Percentage')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
position_counts = df['Position'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(position_counts.index, position_counts.values)
plt.title('Player Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Number of Players')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
position_team_counts = df.groupby(['Position', 'Team'])['Name'].count().reset_index()
positions = position_team_counts['Position'].unique()
teams = position_team_counts['Team'].unique()
bar_width = 0.35
index = np.arange(len(positions))
fig, ax = plt.subplots(figsize=(12, 6))
for i, team in enumerate(teams):
    team_data = position_team_counts[position_team_counts['Team'] == team]
    team_positions = team_data['Position'].values
    team_counts = team_data['Name'].values
    position_indices = np.where(np.isin(positions, team_positions))[0]
    ax.bar(index[position_indices] + i * bar_width, team_counts, bar_width, label=team)
ax.set_xlabel('Position')
ax.set_ylabel('Number of Players')
ax.set_title('Player Distribution by Position and Team')
ax.set_xticks(index + bar_width * (len(teams) -1)/ 2)
ax.set_xticklabels(positions, rotation=45, ha='right')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
age_bins = [0, 25, 30, 35, 40]
age_labels = ['<25', '25-29', '30-34', '35-39']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
age_group_counts = df['Age Group'].value_counts()
predominant_age_group = age_group_counts.idxmax()
print(f"The predominant age group among players is: {predominant_age_group}")
plt.figure(figsize=(10, 6))
plt.bar(age_group_counts.index, age_group_counts.values)
plt.title('Player Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Players')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.bar(predominant_age_group, age_group_counts[predominant_age_group], color='red')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
age_bins = [0, 25, 30, 35, 40]
age_labels = ['<25', '25-29', '30-34', '35-39']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
team_age_counts = df.groupby(['Team', 'Age Group'])['Name'].count().reset_index()
predominant_age_groups = team_age_counts.loc[team_age_counts.groupby('Team')['Name'].idxmax()]
print("Predominant Age Groups by Team:")
print(predominant_age_groups[['Team', 'Age Group']])
teams = predominant_age_groups['Team'].unique()
age_groups = predominant_age_groups['Age Group'].unique()
fig, ax = plt.subplots(figsize=(12, 6))
x = np.arange(len(teams))
bar_width = 0.8
for i, age_group in enumerate(age_groups):
    team_data = predominant_age_groups[predominant_age_groups['Age Group'] == age_group]
    ax.bar(x[np.where(np.isin(teams, team_data['Team']))[0]], team_data['Name'],
            bar_width, label=age_group)
ax.set_xlabel('Team')
ax.set_ylabel('Number of Players in Predominant Age Group')
ax.set_title('Predominant Age Group by Team')
ax.set_xticks(x)
ax.set_xticklabels(teams, rotation=45, ha='right')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
age_bins = [0, 25, 30, 35, 40]
age_labels = ['<25', '25-29', '30-34', '35-39']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
plt.figure(figsize=(15, 8))
sns.countplot(data=df, x='Team', hue='Age Group')  
plt.xticks(rotation=45, ha='right')
plt.title('Age-Based Player Distribution by Team')
plt.xlabel('Team')
plt.ylabel('Number of Players')
plt.tight_layout()  
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
team_position_salary = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
max_salary_team_position = team_position_salary.loc[team_position_salary['Salary'].idxmax()]
print("Team and Position with Highest Salary Expenditure:")
print(f"Team: {max_salary_team_position['Team']}")
print(f"Position: {max_salary_team_position['Position']}")
print(f"Total Salary Expenditure: {max_salary_team_position['Salary']}")
plt.figure(figsize=(15, 8))
heatmap_data = team_position_salary.pivot(index="Team", columns="Position", values="Salary")
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="viridis")
plt.title('Total Salary Expenditure by Team and Position')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
correlation = df['Age'].corr(df['Salary']) #finding the Correlation Coefficient
print(f"Correlation between Age and Salary: {correlation}")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Salary')
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.tight_layout()  
plt.show()
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Age', y='Salary')
plt.title('Correlation between Age and Salary (with Trend Line)')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.tight_layout()  
plt.show()
```
